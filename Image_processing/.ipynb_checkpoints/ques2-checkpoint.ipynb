```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import seaborn as sns
```

```python
#reading
ima = mpimg.imread('airplane.tiff')
n,m = ima.shape
```

```python
arr = np.ones((n,m))
pixel_max = 0
for i in range (0,n):
    for j in range(0,m):
        arr[i][j] = ima[i][j]
        pixel_max = max(pixel_max,arr[i][j])

arr = arr.astype(int)
ima.shape
```

```python
def get_bitplane(arr,i,n,m):
    ans = np.ones((n,m))
    ans = ans.astype(int)
    a = 1 << i
    for i in range(0,n):
        for j in range(0,m):
            ans[i][j] = (arr[i][j]/a)%2
                
    return ans

def qunati_los(arr,i,n,m):
    ans = np.ones((n,m))
    ans = ans.astype(int)
    a = 1<<i
    for i in range(0,n):
        for j in range(0,m):
            ans[i][j] = arr[i][j]/2 
            
    return ans
```

```python
#Bitplanes
for i in range(0,8):
    plt.figure()
    a = get_bitplane(arr,i,n,m)
    plt.imshow(a,cmap='gray')    
```

```python
plt.imshow(ima,cmap='gray')
```

```python
#qunatization
a = arr;
i = 0
while i < 8:
    plt.figure()
    a = qunati_los(a,i,n,m)
    plt.imshow(a,cmap='gray')
    i = i+1
```

```python
his = np.zeros(256)
his = his.astype(int)
for i in range(0,n):
    for j in range(0,m):
        his[arr[i][j]] = his[arr[i][j]]+1
```

```python
# Histogram
X = []
Y = []
for i in range(0,256):
    X.append(i)
    Y.append(his[i])
    
plt.plot(X,Y)
```
