# Question 2:

### Functions Description
<p>point_operator(image,n,m,p,l): Performs  simple linear point operations</p>
<p>log_range_compression(image,n,m): . Log magnitude compression implementation </p>
<p>FSCS(ima,n,m) : uses all remaining functions for flatteing histogram</p>

```python
def point_operator(ima,n,m,p,l):
    J = np.zeros((n,m))
    for i in range(0,n):
        for j in range(0,m):
            J[i][j] = round(p*ima[i][j]+l)
            if J[i][j] > 255:
                J[i][j] = 255
            if J[i][j] < 0:
                J[i][j] = 0
    J = J.astype(int)
    return J


def log_range_compression(ima,n,m):
    J = np.zeros((n,m))
    for i in range(0,n):
        for j in range(0,m):
            J[i][j] = round(math.log(1+ima[i][j],2))
            if J[i][j] > 255:
                J[i][j] = 255
            if J[i][j] < 0:
                J[i][j] = 0
    J = J.astype(int)
    return J

def get_probabilities(arr,n,m):
    count = n*m
    ans = np.zeros(256)
    for i in range(0,256):
        ans[i] = arr[i]/count
    return ans

def get_cummulative_freq(probability_arr,n,m):
    for i in range(1,256):
        probability_arr[i] = probability_arr[i]+probability_arr[i-1]

    return probability_arr

def get_flatten(cummulative_freq):
    for i in range(0,256):
        cummulative_freq[i] = round(cummulative_freq[i]*255)
    return cummulative_freq

def get_image_flatten(ima,flatten,n,m):
    ans = np.zeros((n,m))
    for i in range(0,n):
        for j in range(0,m):
            ans[i][j] = flatten[ima[i][j]]
            
    ans = ans.astype(int)
    return ans

def FSCS(ima,n,m):
    maxd = np.amax(ima)
    mind = np.amin(ima)
    p = float(255/float(maxd-mind))
    l = -float((255*mind)/float(maxd-mind))
    J = point_operator(ima,n,m,p,l)
    return J
    
```

```python
#Flattening
def image_flattening(ima,n,m):    
    arr = get_frequency_vector(ima,n,m)
    probability_arr = get_probabilities(arr,n,m)
    cummulative_freq = get_cummulative_freq(probability_arr,n,m)
    flatten = get_flatten(cummulative_freq)
    new_ima = get_image_flatten(ima,flatten,n,m)
    return new_ima

flatten_histogram_image = image_flattening(ima,n,m)
fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.hist(ima.ravel(),256,[0,256])
ax2.hist(flatten_histogram_image.ravel(),256,[0,256])
print("BEFORE","AFTER")
```

# Question 3:

###  Functions Description:
<p> nearest_neighbour(image,scale,scale): Zooms images using nearest_neighbour method </p>
    <p> bilinear: helps in finding A,B,C,D while doing bilinear transformation(linear_interpolation+bilinear_interpolation does the same ) </p>
    <p> bilinear_interpolation_func(image,fx,fy): Zooms images using Bilinear_interpolation method</p>

```python

def nearest_neighbour(image,fx,fy):
    n,m = image.shape
    
    newN = int(n*float(fx))
    newM = int(m*float(fy))
    
    ratioN = n/newN
    ratioM = m/newM
    
    new_image = np.zeros((newN,newM),dtype = np.uint8)
    
    for i in range(new_image.shape[0]):
        for j in range(new_image.shape[1]):
            
            x = round(ratioN*i)
            y = round(ratioM*j)
            
            x = x-1 if x == n else x
            y = y-1 if y == m else y
            
            temp = image[x][y]
            new_image[i][j] = temp
            
    return new_image


def bilinear1(pt1,pt2,pt3,pt4,pt,i,j):
    matix = np.array([[1,pt1[0],pt1[1],pt1[0]*pt1[1]],
                     [1,pt2[0],pt2[1],pt2[0]*pt2[1]],
                     [1,pt3[0],pt3[1],pt3[0]*pt3[1]],
                     [1,pt4[0],pt4[1],pt4[0]*pt4[1]],])
    right_side = np.array([[pt1[2]],
                          [pt2[2]],
                          [pt3[2]],
                          [pt4[2]]])

    inverse = np.linalg.pinv(matix)
    ans = inverse@right_side
    P = ans[0][0]+ans[1][0]*i+ans[2][0]*j+ans[3][0]*i*j
    return P

def linear_interpolation(pt1, pt2, unknown):
        temp1 = pt1[1]*((pt2[0]-unknown[0])/(pt2[0]-pt1[0]))
        temp2 = pt2[1]*((unknown[0]-pt1[0])/(pt2[0]-pt1[0]))

        return temp1+temp2

def bilinear_interpolation(pt1, pt2, pt3, pt4, unknown):

        if pt1[0] == pt3[0]:
            return linear_interpolation((pt1[1], pt1[2]), (pt2[1], pt2[2]), (unknown[1], unknown[2]))

        if pt1[1] == pt2[1]:
            return linear_interpolation((pt1[0], pt1[2]), (pt3[0], pt3[2]), (unknown[0], unknown[2]))

        R1 = linear_interpolation((pt1[0], pt1[2]), (pt3[0], pt3[2]), (unknown[0], unknown[2]))
        R2 = linear_interpolation((pt2[0], pt2[2]), (pt4[0], pt4[2]), (unknown[0], unknown[2]))
        P = linear_interpolation((pt1[1], R1), (pt2[1], R2), (unknown[1], unknown[2]))

        return P

def bilinear_interpolation_func(image,fx,fy):
    n,m = image.shape
    
    newN = int(n*float(fx))
    newM = int(m*float(fy))
    
    ratioN = n/newN
    ratioM = m/newM
    
    new_image = np.ones((newN,newM))
    
    for i in range(new_image.shape[0]):
        for j in range(new_image.shape[1]):
            
            x = ratioN*i
            y = ratioM*j
            
            x1 = math.floor(x)
            x2 = math.ceil(x)
            y1 = math.floor(y)
            y2 = math.ceil(y)
            
            x1 = x1-1 if x1 == n else x1
            x2 = x2-1 if x2 == n else x2
            y1 = y1-1 if y1 == m else y1
            y2 = y2-1 if y2 == m else y2
            
            pt1 = (x1, y1, image[x1][y1])
            pt2 = (x1, y2, image[x1][y2])
            pt3 = (x2, y1, image[x2][y1])
            pt4 = (x2, y2, image[x2][y2])
            
            unknown = (x,y,0)
            if pt1[0] == pt3[0] and pt1[2] == pt2[2]:
                new_image[i][j] = image[int(x)][int(y)]
            else:
                new_image[i][j] = bilinear_interpolation(pt1, pt2, pt3, pt4, unknown)
                
    return new_image
    
```
