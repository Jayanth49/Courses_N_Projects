```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import seaborn as sns
```

```python
#reading image
ima = mpimg.imread('airplane.tiff')
plt.imshow(ima,cmap='gray')
ima.shape
```

```python
n,m = ima.shape
```

```python
# Getting frequency's of each scale
arr = np.zeros(256)
for i in range(0,n):
    for j in range(0,m):
        arr[ima[i][j]] = arr[ima[i][j]]+1
        
arr = arr.astype(int)
```

```python
#Histogram.....
X = []
Y = []
for i in range(0,256):
    X.append(i)
    Y.append(arr[i])
    
plt.plot(X,Y)
```

```python
# function to get maximum frequency_ index in frequecy array
def get_max_index(arr):
    n = arr.size
    count = -1
    index = -1
    for i in range(0,n):
        if arr[i] > count:
            count = arr[i]
            index = i
    return index

#function to convert into binary
def convert_into_binary(ima,t,t1):
    n,m = ima.shape
    ans = np.zeros((n,m))
    for i in range(0,n):
        for j in  range(0,m):
            if ima[i][j] < t and ima[i][j] > t1:
                ans[i][j] = 1
            else:
                ans[i][j] = 0
    return ans

# function to reverse all bits(0's to 1's) (viceversa)
def reversing(arr):
    n,m = arr.shape
    ans = np.zeros((n,m))
    for i in range(0,n):
        for j in range(0,m):
            if arr[i][j] == 0:
                ans[i][j] = 1
            else:
                ans[i][j] = 0
    return ans

# to get region_counts_frequency
def get_region_counts(arr):
    arr = arr.astype(int)
    region_counts = np.zeros((n*m))
    for i in range(0,n):
        for j in range(0,m):
            region_counts[arr[i][j]] = region_counts[arr[i][j]]+1
    return region_counts

#to get max_frequency region
def get_max_index(region_counts):
    region_counts = region_counts.astype(int)
    siz = region_counts.size
    count_m = 0
    index = -1
    for i in range(0,siz):
        if count_m < region_counts[i]:
            count_m = region_counts[i]
            index = i
    return index
```

```python
#Binary_Image
threshold = get_max_index(arr)
print(threshold)
binar_ima = convert_into_binary(ima,threshold+25,threshold-20)
plt.imshow(binar_ima,cmap='gray')
# binary_r  = reversing(binar_ima)
# plt.imshow(binary_r,cmap=plt.cm.gray)
# binar_ima.shape
```

```python
#function that helps in finding Largest Connected Component
def blob_removal(arr):
    n,m = arr.shape
    print(n,m)
    visited = np.zeros((n,m))
    region_index = np.zeros((n,m))
    p = 1
    ax = [1,-1,0,0]
    ay = [0,0,1,-1]
    q = []
    for i in range(0,n):
        for j in range(0,m):
            if arr[i][j] == 0 and visited[i][j] == 0:
                q.append((i,j))
                visited[i][j] = 1
                region_index[i][j] = p
                p = p+1
                while len(q):
                    x,y = q.pop(0)
                    for k in range(0,4):
                        nx = x+ax[k]
                        ny = y+ay[k]
                        if nx >= 0 and nx < n and ny >= 0 and ny < m and visited[nx][ny] == 0 and arr[nx][ny] == 0:
                            q.append((nx,ny))
                            visited[nx][ny] = 1
                            region_index[nx][ny] = region_index[x][y]
                
                    
    return region_index
```

```python
#Here(next 2 cells) we finding largest_connected  component
brr = blob_removal(binar_ima)
brr = brr.astype(int)
n,m = brr.shape
c = 0
region_counts = np.zeros((n*m))
for i in range(0,n):
    for j in range(0,m):
        region_counts[brr[i][j]] = region_counts[brr[i][j]]+1

region_counts
sum(region_counts)
# print(n,m)
```

```python
region_counts = region_counts.astype(int)
siz = region_counts.size

def get_indicis(region_counts,n):
    indices = (-region_counts).argsort()[:n]
    return indices
```

```python
#Image After first blob removal
indices = get_indicis(region_counts,2)
n,m = ima.shape
for i in range(0,n):
    for j in range(0,m):
        if brr[i][j] in indices and brr[i][j] !=0:
            brr[i][j] = 0
        else:
            brr[i][j] = 1
plt.imshow(brr,cmap=plt.cm.gray)
```

```python
#Doing Blob removal for reversed Image
again_brr = reversing(brr)
plt.imshow(again_brr,cmap=plt.cm.gray)
new_blob = blob_removal(again_brr)
```

```python
#Final Image...
region_counts = get_region_counts(new_blob)
index = get_max_index(region_counts)
print(index)
final_answer = again_brr
for i in range(0,n):
    for j in range(0,m):
        if new_blob[i][j] != index :
            final_answer[i][j] = 1
            
plt.imshow(final_answer,cmap=plt.cm.gray)
```
