```python
# mu says position
#cluster_id says cluster to which it belongs

def initilize_mu(k):
    np.random.seed(48)
    return np.random.randint(low = 0, high = data.shape[1],size = (k,2))

def calculate(temp):
    return np.sqrt(np.sum(temp**2,axis=1))

def cost_function(mu,cluster_id,data):
    cost = 0
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            cost += np.sum(((mu[cluster_id[i][j]]-data[i][j])**2))

    return cost

def find_index(temp):
    n = temp.size
    minElement = np.amin(temp)
    index = -1
    for i in range(n):
        if temp[i] == minElement:
            index = i
            break
            
    return index


def k_means_clustering(data,k,max_iters = 1000):
    print("Function.....")
    mu = initilize_mu(k)
    cluster_id = np.zeros((data.shape[0],data.shape[1]),dtype=int)
                   
    num_iters = 0
    error = 1000
    
    mu_data = np.zeros((k,3))

    for i in range(k):
        for j in range(3):
            mu_data[i][j] = data[mu[i][0]][mu[i][1]][j]
    
    cost_values = []

            
    while (num_iters < max_iters) and (error > 1) : 
#         print("Enters while loop.....")

        print("Iteration no: ",num_iters)
        for i in range(data.shape[0]):
            for j in range(data.shape[1]):
                temp = mu_data-data[i][j]
                #calculate length
                length = calculate(temp)
                #find_min_index
                index = find_index(length)
                cluster_id[i][j] = index
    
    
        new_mu_data = np.zeros((k,3))
        cluster_size = np.zeros(k)
        for i in range(data.shape[0]):
            for j in range(data.shape[1]):
                cluster_size[cluster_id[i][j]] += 1
                for l in range(3): 
                    new_mu_data[cluster_id[i][j]][l] += data[i][j][l]
                    
                    
        for i in range(k):
            new_mu_data[i] = new_mu_data[i]/cluster_size[i]
        
        J = cost_function(mu_data,cluster_id,data)
        cost_values.append(J)
        error = np.sum(np.abs(mu_data-new_mu_data))
        mu_data = new_mu_data
#         print(num_iters)
        num_iters += 1
        print("error:",error)
        print("cost:",J)


    return mu_data,cluster_id,cost_values,num_iters
            
```

```python
k = 4
mu,cluster_id,cost,num_iters = k_means_clustering(data,k)
```

```python
# data_ready->64*64

def PCA_to_decorrelate(data):
    data_Standarised = data
    for i in range(data.shape[0]):
        data_Standarised[i] = data[i]-np.mean(data,axis = 1)[i]
        
    Cxx = (data_Standarised@data_Standarised.T)/data_Standarised.shape[0]
    eig_vals,eig_vect = np.linalg.eig(Cxx)
    eig_v = np.real(eig_vect)
    
    P = eig_v.T
    Y = P@data_Standarised
    
    return Y


def Verifying_PCA(Y):
    Cyy = (Y@Y.T)/data.shape[0]
    Cyy_round = np.round(Cyy,decimals = 6)
    
    for i in range(Cyy_round.shape[0]):
        for j in range(Cyy_round.shape[1]):
            if (i!= j)  and (Cyy_round[i][j] != 0):
                return False
    return True
```

### Examples Where PCA Fails

<p>1) PCA has Assumptions Linearity,Large variance as structures and Principle components are perpendicular. </p>
<p>2) So, When these assumptions are failed our PCA also fails </p>
<p> SO PCA May fail in these cases:
<ol>
    <li>When we get non linear data</li>
    <li> When data is highly biased(like blobs) </li>
    <li>Principle components need not be perpendicular</li>
</ol>

### Finding Maximum Likelihood Estimates

```python
#Binomial
# n,p = 5,0.4
# s = np.random.binomial(n, p, 1000)

def binomial_parameter_estimate(n,p,num_samples):
    data = np.random.binomial(n, p,size = num_samples)
    estimated_p = np.sum(data)/(n*num_samples)
    return estimated_p

#poisson
# s = np.random.poisson(lam=5, size=1000)

def poissson_paramter_estimate(lamb,num_samples):
    data = np.random.poisson(lamb,size = num_samples)
    estimated_lambda = np.sum(data)/num_samples
    return estimated_lambda

#Exponential(1/l)
# s = np.random.exponential(scale = 5,size = 1000)

def exponential_paramter_estimation(lamb,num_samples):
    data = np.random.exponential(scale = lamb,size = num_samples)
    estimated_lambda = (np.sum(data))/num_samples
    return estimated_lambda

#Gaussian
# s = np.random.normal(loc = 0.0,scale = 5,size = 1000)

def gaussian_parameters_estimation(mean,std,num_samples):
    data = np.random.normal(loc = mean,scale = std,size = num_samples)
    estimated_mean = np.sum(data)/num_samples
    estimated_variance = np.sum((data-estimated_mean)**2)/num_samples
    return estimated_mean,estimated_variance

#Laplacian
# s = np.random.laplace(loc = 0.0,scale = 5,size = 1000)

def Laplace_parameters_estimation(mean,b,num_samples):
    data = np.random.laplace(loc = mean,scale = b,size = num_samples)
    estimated_mean = np.median(data)
    estimated_lambda = np.sum(np.abs(data-estimated_mean))/num_samples
    return estimated_mean,estimated_lambda
```

### Comparing Estimates with original Parameters with changing n from 1 to pow(5,8)

```python
#Binomial

def compare_binomial_estimates(original_p):
    n = 1
    barWidth = 0.25
    h1 = []
    h2 = []


    for i in range(8):
        estimated_p = binomial_parameter_estimate(10,original_p,n)
    #     print(original_p-estimated_p)
        h1.append(original_p)
        h2.append(estimated_p)
        n = n*5

    r1 = np.arange(len(h1))
    r2 = [x + barWidth for x in r1]

    plt.ylim(0,original_p+0.3)

    # plt.bar(x,h,color='#7f6d5f')
    plt.bar(r1, h1, color='#2d7f5e', width=barWidth, edgecolor='white', label='Original_p')
    plt.bar(r2, h2, color='#557f2d', width=barWidth, edgecolor='white', label='Estimated_p')
    plt.xlabel('number of samples = Power(5,x)', fontweight='bold')
    plt.ylabel('values',fontweight='bold')
    plt.legend()
    plt.show()


#poisson
def compare_poisson_estimates(original_lambda):
    n = 1
    barWidth = 0.25
    h1 = []
    h2 = []


    for i in range(8):
        np.random.seed(11)
        estimated_lambda = poissson_paramter_estimate(original_lambda,n)
    #     print(original_p-estimated_p)
        h1.append(original_lambda)
        h2.append(estimated_lambda)
        n = n*5

    r1 = np.arange(len(h1))
    r2 = [x + barWidth for x in r1]

    plt.ylim(0,8)

    # plt.bar(x,h,color='#7f6d5f')
    plt.bar(r1, h1, color='#2d7f5e', width=barWidth, edgecolor='white', label='Original_lambda')
    plt.bar(r2, h2, color='#557f2d', width=barWidth, edgecolor='white', label='Estimated_lambda')
    plt.xlabel('number of samples = Power(5,x)', fontweight='bold')
    plt.ylabel('values',fontweight='bold')
    plt.legend()
    plt.show()
    
    
# Exponential
def compare_exponential_estimates(original_lambda):

    n = 1
    barWidth = 0.25
    h1 = []
    h2 = []


    for i in range(8):
        np.random.seed(11)
        estimated_lambda = exponential_paramter_estimation(original_lambda,n)
    #     print(original_lambda-(estimated_lambda))
        h1.append(original_lambda)
        h2.append(estimated_lambda)
        n = n*5

    r1 = np.arange(len(h1))
    r2 = [x + barWidth for x in r1]

    plt.ylim(0,8)

    # plt.bar(x,h,color='#7f6d5f')
    plt.bar(r1, h1, color='#2d7f5e', width=barWidth, edgecolor='white', label='Original_lambda')
    plt.bar(r2, h2, color='#557f2d', width=barWidth, edgecolor='white', label='Estimated_lambda')
    plt.xlabel('number of samples = Power(5,x)', fontweight='bold')
    plt.ylabel('values',fontweight='bold')
    plt.legend()
    plt.show()
    
    
#Gaussain
def compare_gaussian_estimates(original_mean,original_variance):
    n = 1
    barWidth = 0.25
    h1 = []
    h2 = []
    h3 = []
    h4 = []


    for i in range(8):
        np.random.seed(11)
        estimated_mean,estimated_variance = gaussian_parameters_estimation(original_mean,np.sqrt(original_variance),n)
        h1.append(original_mean)
        h2.append(estimated_mean)
        h3.append(original_variance)
        h4.append(estimated_variance)
        n = n*5

    r1 = np.arange(len(h1))
    r2 = [x + barWidth for x in r1]
    r3 = np.arange(len(h3))
    r4 = [x+barWidth for x in r3]

    # plt.ylim(0,10)
    fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(14,10))

    # plt.bar(x,h,color='#7f6d5f')
    ax1.bar(r1, h1, color='#2d7f5e', width=barWidth, edgecolor='white', label='Original_mean')
    ax1.bar(r2, h2, color='#557f2d', width=barWidth, edgecolor='white', label='Estimated_mean')

    ax1.set_title('mean')
    ax1.set_ylabel('value')
    ax1.set_xlabel('Number of Samples = power(5,x)')
    ax1.legend()

    ax2.bar(r3, h3, color='#2d7f5e', width=barWidth, edgecolor='white', label='Original_variance')
    ax2.bar(r4, h4, color='#557f2d', width=barWidth, edgecolor='white', label='Estimated_variance')

    ax2.set_title('variance')
    ax2.set_ylabel('value')
    ax2.set_xlabel('Number of Samples = power(5,x)')
    ax2.legend()
    # fig.legend()
    # fig.show()
    
    
#Laplacain
def compare_laplace_estimates(original_mean,original_b):
    n = 1
    barWidth = 0.25
    h1 = []
    h2 = []
    h3 = []
    h4 = []



    for i in range(8):
        np.random.seed(11)
        estimated_mean,estimated_b = Laplace_parameters_estimation(original_mean,original_b,n)
        h1.append(original_mean)
        h2.append(estimated_mean)
        h3.append(original_b)
        h4.append(estimated_b)
        n = n*5

    r1 = np.arange(len(h1))
    r2 = [x + barWidth for x in r1]
    r3 = np.arange(len(h3))
    r4 = [x+barWidth for x in r3]

    # plt.ylim(0,10)
    fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(14,10))

    # plt.bar(x,h,color='#7f6d5f')
    ax1.bar(r1, h1, color='#2d7f5e', width=barWidth, edgecolor='white', label='Original_mean')
    ax1.bar(r2, h2, color='#557f2d', width=barWidth, edgecolor='white', label='Estimated_mean')

    ax1.set_title('mean')
    ax1.set_ylabel('value')
    ax1.set_xlabel('Number of Samples = power(5,x)')
    ax1.legend()

    ax2.bar(r3, h3, color='#2d7f5e', width=barWidth, edgecolor='white', label='Original_b')
    ax2.bar(r4, h4, color='#557f2d', width=barWidth, edgecolor='white', label='Estimated_b')

    ax2.set_title('b')
    ax2.set_ylabel('value')
    ax2.set_xlabel('Number of Samples = power(5,x)')
    ax2.legend()
```

#### Comparing Distributions---generating samples with original parameters and estimated parameters

```python
#binomial

def compare_samples_from_estimated_data_binomial(n,p,num_samples):
    original_data = np.random.binomial(n, p,size = num_samples)
    estimated_p = binomial_parameter_estimate(n,original_p,num_samples)
    estimated_data = np.random.binomial(n,estimated_p,size = num_samples)

    bins = np.linspace(-10, 50, 100)

    plt.hist(original_data, bins, alpha=0.5, label='Original_data')
    plt.hist(estimated_data, bins, alpha=0.5, label='estimated_data')
    plt.legend(loc='upper right')
    plt.show()
    
#Poisson

def compare_samples_from_estimated_data_poisson(original_lambda,num_samples):
    original_lambda = 5
    num_samples = 1000
    original_data = np.random.poisson(original_lambda,size = num_samples)
    estimated_lambda = poissson_paramter_estimate(original_lambda,num_samples)
    estimated_data = np.random.poisson(estimated_lambda,num_samples)

    bins = np.linspace(-10, 30, 100)

    plt.hist(original_data, bins, alpha=0.5, label='Original_data')
    plt.hist(estimated_data, bins, alpha=0.5, label='estimated_data')
    plt.legend(loc='upper right')
    plt.show()
    
# Exponential
def compare_samples_from_estimated_data_exponential(original_lambda,num_samples):

    original_data = np.random.exponential(original_lambda,size = num_samples)
    estimated_lambda = exponential_paramter_estimation(original_lambda,num_samples)
    estimated_data = np.random.exponential(estimated_lambda,size = num_samples)

    bins = np.linspace(-10, 30, 100)

    plt.hist(original_data, bins, alpha=0.5, label='Original_data')
    plt.hist(estimated_data, bins, alpha=0.5, label='estimated_data')
    plt.legend(loc='upper right')
    plt.show()
    
#Gaussain
def compare_samples_from_estimated_data_gaussian(original_mean,original_variance,num_samples):
    original_data = np.random.normal(original_mean,np.sqrt(original_variance),size = num_samples)
    estimated_mean,estimated_variance = gaussian_parameters_estimation(original_mean,np.sqrt(original_variance),num_samples)
    estimated_data = np.random.normal(estimated_mean,np.sqrt(estimated_variance),size = num_samples)
    bins = np.linspace(-10, 30, 100)

    plt.hist(original_data, bins, alpha=0.5, label='Original_data')
    plt.hist(estimated_data, bins, alpha=0.5, label='estimated_data')
    plt.legend(loc='upper right')
    plt.show()
    
#Lapalce
def compare_samples_from_estimated_data_lapalce(original_mean,original_b,num_samples):
    original_data = np.random.laplace(original_mean,original_b,size = num_samples)
    estimated_mean,estimated_b = Laplace_parameters_estimation(original_mean,original_b,num_samples)
    estimated_data = np.random.laplace(estimated_mean,estimated_b,size = num_samples)
    bins = np.linspace(-10, 30, 100)

    plt.hist(original_data, bins, alpha=0.5, label='Original_data')
    plt.hist(estimated_data, bins, alpha=0.5, label='estimated_data')
    plt.legend(loc='upper right')
    plt.show()
```
